```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

###Load in the data set and check the first few rows

```python
df = pd.read_csv('/content/Zolar Case.csv')
df.head()
```

###Check the data types

```python
df.info()
```

###Transfrom date strings to datetime format

```python
df['Date when order started status'] = pd.to_datetime(df['Date when order started status'])
df['Date when order left status'] = pd.to_datetime(df['Date when order left status'])
df.info()
```

##Approximate installation dates for status 21

###Create anew data frame with only status 21

```python
status_21 = df[df['Status Number']==21.0]
status_21.head()
status_21.info()
```

###Check the number of duplicate order numbers

```python
len(status_21['Order Number'])-len(status_21['Order Number'].drop_duplicates())
```

###Create a new column that has the duration of each entry

```python
status_21['Duration'] = (status_21['Date when order left status'] - status_21['Date when order started status']).astype('timedelta64[h]')
status_21.head()
```

###Drop null values so that I get a clear picture of mean and average (not needed!)

```python
status_21_nonull = status_21.dropna()
status_21_nonull.info()
```

```python
status_21_nonull['Duration'].mean()
```

```python
status_21_nonull['Duration'].median()
```

Choosing to go with average because we have 83 entries that have multiple 21 entries that skew the median because they are so short

Calculate the max date

```python
Max_date = status_21['Date when order started status'].max()
Max_date
```

###Fill the null values in the data frame with the average duration + the latest date

```python
status_21['Date when order left status'] = status_21['Date when order left status'].fillna(Max_date + pd.Timedelta(hours=202.79))
status_21
```

###Displaying the Null values in anew data frame 

This gives me the final view of the unfinished installations that have been filled with a new value

```python
status_21_nulls = status_21[status_21.isnull().any(axis=1)]
status_21_nulls
```

Save in new CSV

```python
status_21_nulls.to_csv('Approximate Installation times.csv')
```
